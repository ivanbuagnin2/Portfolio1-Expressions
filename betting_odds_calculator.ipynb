```python
print("================================")
print(" BASKETBALL PARLAY CALCULATOR ")
print("================================")

# User input
player1 = input("Enter Player 1: ")
odds1 = float(input("Enter Odds for Player 1: "))

player2 = input("Enter Player 2: ")
odds2 = float(input("Enter Odds for Player 2: "))

player3 = input("Enter Player 3: ")
odds3 = float(input("Enter Odds for Player 3: "))

bet_amount = float(input("Enter Bet Amount (PHP): "))

# Computations
combined_odds = odds1 * odds2 * odds3
possible_payout = combined_odds * bet_amount
possible_profit = possible_payout - bet_amount

# Output
print("\n========== PARLAY SUMMARY ==========")

print("Player 1:", player1)
print("Odds 1:", odds1)

print("Player 2:", player2)
print("Odds 2:", odds2)

print("Player 3:", player3)
print("Odds 3:", odds3)

print("\nCombined Odds:", round(combined_odds, 2))
print("Possible Payout: PHP", round(possible_payout, 2))
print("Possible Profit: PHP", round(possible_profit, 2))
```
